```python
import requests

#Read in the dataframe

df = pd.read_csv('data.csv')



#Create empty columns for lat and lon

df['lat'] = ""

df['lon'] = ""



#Loop through the rows of the dataframe, use area code to query OSM API and save lat and lon in the respective columns  

for index, row in df.iterrows():



    area_code = row['area_code']



    #Query OSM API with area code  

    response = requests.get('https://nominatim.openstreetmap.org/search?q=' + area_code + '&format=json')



    #Save lat and lon from response to respective columns  

    df.loc[index, 'lat'] = response.json()[0]['lat']

    df.loc[index, 'lon'] = response.json()[0]['lon']



 #Save the dataframe with new columns  

df.to_csv('data_with_coordinates.csv', index=False)
```
